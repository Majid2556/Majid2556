```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
raw_data = pd.read_excel('Exam.xlsx')
raw_data
```

```python
df = raw_data.copy()

df['Accepted']=df['Accepted'].map({'Yes':1, 'No':0})
df['Gender']=df['Gender'].map({'Male':1, 'Female':0})

df
```

```python
x = df.drop('Accepted',axis=1)
y = df.Accepted
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(x,y)
```

```python
x_test_raw = pd.read_excel('Test.xlsx')
x_test = x_test_raw.drop('Accepted',axis=1)

x_test['Gender'] = x_test['Gender'].map({'Male':1, 'Female':0})
y_test= x_test_raw['Accepted'].map({'Yes':1, 'No':0})

x_test
```

```python
y_pred = logistic_regression.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)

accuracy_percent = 100*accuracy

accuracy_percent, round(accuracy,4)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```
