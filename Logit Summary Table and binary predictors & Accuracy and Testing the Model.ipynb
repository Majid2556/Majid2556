```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_excel('Exam.xlsx')
raw_data
```

```python
data = raw_data.copy()
data['Accepted'] = data['Accepted'].map({'Yes':1,'No':0})
data['Gender'] = data['Gender'].map({'Male':1, 'Female':0})
data
```

## Logit Regression with different independent variables

```python
y = data['Accepted']
x0 = np.ones(151)

reg_log0 = sm.Logit(y,x0).fit()
reg_log0.summary()
```

```python
y = data['Accepted']
x1 = data ['Score']

x = sm.add_constant(x1)
reg_log1 = sm.Logit(y,x).fit()
reg_log1.summary()
```

```python
y = data['Accepted']
x2 = data[['Score','Gender']]

x_1= sm.add_constant(x2)
reg_log2 = sm.Logit(y,x_1).fit()
reg_log2.summary()
```

```python
y = data['Accepted']
x1 = data ['Gender']

x = sm.add_constant(x1)
reg_log1 = sm.Logit(y,x).fit()
reg_log1.summary()
```

### Due to this block above, we can find that gender feature is not very effective (it's actually not suitable for our model)

## Testing Our Model

```python
test = pd.read_excel('Test.xlsx')
test
```

```python
test['Accepted'] = test['Accepted'].map({'Yes': 1, 'No': 0})

test['Gender'] = test['Gender'].map({'Female': 1, 'Male': 0})

test
```

```python
test_actual = test['Accepted']
test_data = test.drop(['Accepted'],axis=1)
test_data = sm.add_constant(test_data)
test_data
```

```python
def confusion_matrix(data,actual_values,model):
    pred_values = model.predict(data)
    bins = np.array([0,0.5,1])
    cm = np.histogram2d(actual_values,pred_values,bins=bins)[0]
    accuracy = round ((cm[0,0]+cm[1,1])/cm.sum(),4)
    return cm,accuracy
```
